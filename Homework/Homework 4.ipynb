# Homework 4 - Newton-Raphson for equations

Find the solution for  
\begin{eqnarray}
x_{1}^2+x_{1}x_{2}=10\\
x_{2}+3x_{1}x_{2}^2=57
\end{eqnarray}

```python
"""
First, I make a quick plot of the two equations equations.
"""
import matplotlib.pyplot as plt
import numpy as np

def f1(x):
    return (10.-x*x)/x

def f2(x):
    return (-1+np.sqrt(1.+12.*x*57.))/6./x

def f3(x):
    return (-1-np.sqrt(1.+12.*x*57.))/6./x
    

x1 = np.linspace(0.1, 10, 100)
y1 = f1(x1)
y2 = f2(x1)
y3 = f3(x1)
y4 = np.full_like(x1, 3)
y5 = np.full_like(x1, -2.11)

plt.figure(figsize=(12, 7))
plt.plot(x1, y1, 'r-')
plt.plot(x1, y2, 'b-')
plt.plot(x1, y3, 'g-')
plt.plot(x1, y4, '--')
plt.plot(x1, y5, '--')
plt.xlabel('x1')
plt.ylabel('x2')
plt.ylim(-10,10)
plt.xlim(0,10)
plt.grid()
plt.annotate(s='Solution 1', xy=(3, 6))
plt.annotate("", xy=(2, 3), xytext=(3, 6), arrowprops=dict(arrowstyle="->"))
plt.annotate(s='Solution 2', xy=(4, -7.5))
plt.annotate("", xy=(4.39, -2.11), xytext=(4, -7.0), arrowprops=dict(arrowstyle="->"))
plt.title("Intersection of two equations: x2 versus x1")

plt.show()
```

The Newton-Raphson technique uses the following procedure:

1) Make an initial guess of the solution, $x_0$.

2) Find a correction so that $f(x + \Delta x) = 0$

Since we are working with a system of equations, we want to solve:

\begin{equation}
f_{i}(x_{1}+\Delta x_{1}, x_{2}+\Delta x_{2})\simeq f_{i}(x_{1},x_{2})+\frac{\partial f_{i}}{\partial x_{1}}\Delta x_{1}+\frac{\partial f_{i}}{\partial x_{2}}\Delta x_{2}=0 \qquad i=1,2
\end{equation}

\begin{equation}
\left[ \begin{array}{c} f_{1}\\ f_{2} \end{array}\right] + \begin{bmatrix} \partial f_{1}/\partial x_{1} & \partial f_{1}/\partial x_{2}  \\ \partial f_{2}/\partial x_{1} & \partial f_{2}/\partial x_{2} \end{bmatrix}\left[ \begin{array}{c} \Delta x_{1} \\ \Delta x_{2}  \end{array}\right] =0 
\end{equation}

So we want to solve the matrix equation
\begin{equation}
F'\Delta {\bf{x}}=-\bf{F}
\end{equation}

In order to find the step size, then, we simply use the following relation we derived in class:

\begin{equation}
 \begin{bmatrix} a & b \\c & d \end{bmatrix}\left[ \begin{array}{c} \Delta x_{1} \\ \Delta x_{2}  \end{array}\right] =-\left[ \begin{array}{c} f_{1}\\ f_{2} \end{array}\right] 
\end{equation}
The solution is
\begin{eqnarray}
\Delta x_1=\frac{bf_2-df_1}{ad-bc}\\
\Delta x_2=\frac{cf_1-af_2}{ad-bc}
\end{eqnarray}

where

\begin{eqnarray}
a = \frac{\partial f_1}{\partial x_1},
b = \frac{\partial f_1}{\partial x_2}, 
c = \frac{\partial f_2}{\partial x_1},
d = \frac{\partial f_2}{\partial x_2}
\end{eqnarray}

3) Make new x: $x = x_0 + \Delta x$

4) Make x the new $x_0$, and repeat until the solution is found and the system of equations has been minimzed.

From the above plot, it looks like the intersection of the two equations are around $x_1$=2.0. Therefore, I will make an initial guess close to this value.

First, I will define the functions I need in this calculation below.

```python
"""
FUNCTION: df_1
PARAMETERS: 1x2 x array carrying x1 and x2 coordinates
DESCRIPTION: Partial derivatives with respect to coordinates x1, x2 for the "f1" function were calculated
analytically. This function stores the values of these partial derivatives inside the
elements of the df array.
RETURN: 1x2 df array carrying two partial derivatives values.
"""
def df_1(x):
    df = np.array([0., 0.])
    df[0] = 2*x[0]+x[1]
    df[1] = x[0]
    return df

"""
FUNCTION: df_2
PARAMETERS: 1x2 x array carrying x1 and x2 coordinates
DESCRIPTION: Partial derivatives with respect to coordinates x1, x2 for the "f2" function were calculated
analytically. This function stores the values of these partial derivatives inside the
elements of the df array.
RETURN: 1x2 df array carrying two partial derivatives values.
"""
def df_2(x):
    df = np.array([0., 0.])
    df[0] = 3*x[1]*x[1]
    df[1] = 1 + 6*x[0]*x[1]
    return df

"""
FUNCTION: f_1
PARAMETERS: 1x2 x array carrying x1 and x2 coordinates
DESCRIPTION: This function calculates the value of the given "f1" function
at the value of the vector x (x1 and x2)
RETURN: the value of the function at the given x coordinates.
"""
def f_1(x):
    return x[0]*x[0]+x[0]*x[1]-10.

"""
FUNCTION: f_1
PARAMETERS: 1x2 x array carrying x1 and x2 coordinates
DESCRIPTION: This function calculates the value of the given "f1" function
at the value of the vector x (x1 and x2)
RETURN: the value of the function at the given x coordinates.
"""
def f_2(x):
    return x[1]+3.*x[0]*x[1]*x[1]-57.

"""
FUNCTION: F
PARAMETERS: 1x2 x array with x1 and x2 coordinates.
DESCRIPTION: Pass in x array containing x1 and x2 coordinates. Calculate the value of the defined functions
f1 and f2 (also defined above)
RETURN: Return an array f which contains [f1, f2] calculated values.
"""
def F(x):
    f = np.array([0., 0.])
    f[0]=x[0]*x[0]+x[0]*x[1]-10.
    f[1]=x[1]+3.*x[0]*x[1]*x[1]-57.
    return f

"""
FUNCTION: newton raphson with back tracking
PARAMETERS: Coordinates x0 as a 1x2 array, step size dx, max iterations Nmax, error/tolerance eps.
DESCRIPTION: 
-Starting in the for loop: calculate the partial derivatives of the two functions
of interest and store in arrays df1 and df2 (1x2). 
-Store the value of each function at the given x0 (initial guess) 
coordinates in Farr 1x2 array.
-Determine the "correction" terms delta_x1 and delta_x2 and store these in dxt array
-Ensure the function values become smaller when introducing correction terms. If function values 
become bigger, cut correction terms in half. (back tracking)
-Set x0 = x0 + dxt
-Check if function values are below the tolerance (eps)
-Repeat until tolerance is met.
RETURN: Coordinates of intersection/solution to the system of equations (x0)
"""
def newtonraphsonbacktracking(x0, dx, Nmax, eps):
    
    for i in range(0,Nmax):        
        df1 = df_1(x0)
        df2 = df_2(x0)
        Farr = F(x0)
        #print("df1, df2 ", df1, df2)
        delta_x1 = (df1[1]*Farr[1] - df2[1]*Farr[0])/(df1[0]*df2[1]-df1[1]*df2[0])
        delta_x2 = (df2[0]*Farr[0] - df1[0]*Farr[1])/(df1[0]*df2[1]-df1[1]*df2[0])
        #print("delta_x1 = ", delta_x1)
        #print("delta_x2 = ", delta_x2)
        dxt = np.array([delta_x1, delta_x2], dtype=float)
        #print("dxt = ", dxt)
        Farr_dxt = F(x0+dxt)
        #print("Farr_dxt = ", Farr_dxt)
        while (abs(Farr_dxt[0]) > abs(Farr[0]) and abs(Farr_dxt[1]) > abs(Farr[1])):
            #print("Whilte loop dxt", dxt)
            dxt=dxt/2.
            Farr_dxt = F(x0+dxt)     
        print('x0, dxt = ', x0,dxt)
        x0 = x0+dxt
        Farr = F(x0)
        if(abs(Farr[0])<eps and abs(Farr[1])<eps):
            print("Root is at",x0," after ",i," iterations, with a precision of ", abs(F(x0)[0]), abs(F(x0)[1]))
            break
    return x0
```

```python
"""
First, I try an initial guess such that x1 and x2 are smaller than the coordinates of the intersection.

"""
a = [1., 1.]
dx = 1.e-2
Nmax = 100
eps = 1e-8
root = newtonraphsonbacktracking(a, dx, Nmax, eps)
```

```python
"""
Next, I try an initial guess such that x1 and x2 are larger than the coordinates of the intersection.
This will let me see how the program does with backtracking and if the second solution can be found.

"""
a = [10., -2.]
dx = 1.e-2
Nmax = 100
eps = 1e-8
root = newtonraphsonbacktracking(a, dx, Nmax, eps)
```
